```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
from sklearn import metrics, neighbors
expected = y_test
predicted = rf.predict(X_test)

conf_matrix = metrics.confusion_matrix(expected, predicted)

print(conf_matrix)
TP = conf_matrix[0][0]
FN = conf_matrix[0][1]
FP = conf_matrix[1][0]
TN = conf_matrix[1][1]

# Sensitivity, hit rate, recall, or true positive rate
TPR = TP/(TP+FN)
# Specificity or true negative rate
TNR = TN/(TN+FP) 
# Precision or positive predictive value
PPV = TP/(TP+FP)
# Negative predictive value
NPV = TN/(TN+FN)
# Fall out or false positive rate
FPR = FP/(FP+TN)
# False negative rate
FNR = FN/(TP+FN)
# False discovery rate
FDR = FP/(TP+FP)

# Overall accuracy
ACC = (TP+TN)/(TP+FP+FN+TN)



print(f'Accuracy: {ACC}, False Negative: {FNR}, Specificity: {TNR}, Sensitivity: {TPR}, Precision: {PPV}, False Positive: {FPR}')
```

```python
edible_df = one_hot_df[one_hot_df['class'] == 0]
edible_df.head()
```

```python
from scipy.stats import linregress
import matplotlib.pyplot as plt
x_values = one_hot_df["class"]
y_values = one_hot_df["stem-height"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Create scatterplot with linear regression
plt.scatter(x_values,y_values)

plt.plot(x_values,regress_values, "r-")
plt.annotate(line_eq,(10,20),fontsize=15,color="red")
plt.xlabel("Stem-Width")
plt.ylabel("Stem Height")
print(f"The Correlation Coefficient is: {rvalue}")
plt.title("Stem-Height vs Stem-Width")
# plt.tight_layout()
plt.grid()
# plt.savefig("North_Temp_vs_Lat.png")
plt.show()
```
