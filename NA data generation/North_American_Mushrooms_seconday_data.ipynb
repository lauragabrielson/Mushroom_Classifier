## Encode and scale data for machine learning

```python
encoded_data = pd.get_dummies(data=data, columns=[
                                                'cap-surface',
                                                'cap-color',
                                                'does-bruise-or-bleed',
                                                'gill-attachment',
                                                'gill-spacing',
                                                'gill-color',
                                                'cap-shape',
                                                'stem-color',
                                                'has-ring',
                                                'ring-type',
                                                'habitat',
                                                'season'])
encoded_data
```

```python
from sklearn.preprocessing import LabelEncoder
  
le = LabelEncoder()
encoded_data['class']= le.fit_transform(encoded_data['class'])
```

```python
encoded_data
```

```python
y = encoded_data["class"]
target_names = ["edible", "poisonous"]

X = encoded_data.drop("class", axis=1)
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X)
```

```python
X_scaled = X_scaler.transform(X)
```

## Import and run neural network

```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import load_model
mushroom_model = load_model("../python/mushroom_trained.h5")
```

```python
y_categorical = to_categorical(y)
```

```python
mushroom_model.summary()
```

```python
model_loss, model_accuracy = mushroom_model.evaluate(
    X_scaled, y_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
import numpy as np
test = np.expand_dims(X_scaled[34591], axis=0)
test.shape
```

```python
# Make a prediction
print(f"Predicted class: {mushroom_model.predict_classes(test)}")
```

## Run my own RF Train/Test

```python
y = encoded_data["class"]
X = encoded_data.drop("class", axis=1)

X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
scaler = MinMaxScaler()
# transform data
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
rf = RandomForestClassifier(n_estimators=50)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
